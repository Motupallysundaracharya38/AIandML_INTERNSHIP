```python
# xyz:750,abc:265,querty:195
s=input()
list1=s.split(",")
list2=[i.split(":")[1]for i in list1]
list3=[]
for num in list2:
    n=int(num)
    rev=num[::-1]
    while str(num)!=rev:
        n+=int(rev)
        rev=str(n)[::-1]
    list3.append(n)
list3.index(max(list3)).spilt(":")
        
```

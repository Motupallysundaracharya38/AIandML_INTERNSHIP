```python
# preparing the input and output columns
x=boston_df[['RM']]# selecting columns a dataframe
y=boston_df[['MEDV']]
x
```

```python
y
```

```python
boston_df['MEDV']
```

```python
#plt.style.available
#plt.style.use
with plt.style.context("grayscale"):
    plt.plot(x,y,"bo")
    plt.xlabel("RM I/P")
    plt.ylabel("MEDV O/P")
    plt.title("scatter plot")
    
```

```python
## usual sklearn import,instantiate ,fit
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x,y)
```

```python
## print intercept and coefficient and coefficient 
print(lm.intercept_)
print(lm.coef_)
```

```python
# y=mx+c
#m value-->coefficient(9.10210898)
#c value----->intercept(-34.67062078)
(9.10210898)*(6.575)+(-34.67062078)
```

# Interpreting the model:
- A unit increase in X(RM)  is associated with a 9.10210898 unit increase in MEDV

```python
## pipeline-->fort we create a model on the data
# second , we need to compare the model (predicted values with actual values)
```

```python
## with help of predict , we are going to find our predictedd values 
y_pred=lm.predict(x)
y_pred
```

```python
# compare the actual with 
y==y_pred
```

```python
## r2_value-->to check the model
from sklearn.metrics import r2_score
r2_score(y,y_pred)
```

```python
## observe the regression line
with sns.axes_style('dark'):
    sns.regplot(x , y)
```

```python
## scatterplot(regression )
## first we are makuing a scatter plot of the orginal data(X and y)
## we need to plot a line for the  predicted values(x,y_pred)
plt.scatter(x,y,color='red')
plt.plot(x,y_pred,color='black')
plt.xlabel("RM I/P")
plt.ylabel("MEDV O/P")
```
